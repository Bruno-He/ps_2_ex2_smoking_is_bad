## Investigate Confounding Variables

1. Check group averages by smoking status  
Look at the mean values of age and salary for smokers vs. non-smokers.
This tells us whether smokers differ systematically from non-smokers in other characteristics.

2. Visualize possible relationships  
Create visual plots (boxplots or histograms) to see if confounders are distributed differently between smokers and non-smokers.

3. Interpretation  
If smokers tend to be older or have lower income, these factors might influence health outcomes â€” meaning smoking might not be the only cause of mortality differences.

```python

import seaborn as sns
import matplotlib.pyplot as plt

# 1. Compare average age and salary by smoking status
confounder_summary = df.groupby("smoker")[["age", "salary"]].mean().reset_index()
print("Average age and salary by smoking status:")
print(confounder_summary)

# 2. Visualize age distribution by smoking status
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="smoker", y="age")
plt.title("Age Distribution by Smoking Status")
plt.xlabel("Smoker")
plt.ylabel("Age")
plt.show()

# 3. Visualize salary distribution by smoking status
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="smoker", y="salary")
plt.title("Salary Distribution by Smoking Status")
plt.xlabel("Smoker")
plt.ylabel("Salary")
plt.show()

# 4. check proportions by gender
gender_smoker = pd.crosstab(df["gender"], df["smoker"], normalize="columns") * 100
print("\nPercentage of gender within smoker groups:")
print(gender_smoker.round(1))
```

# Conclusion

Smokers appear to have higher survival rate, but this is misleading.

From the age distribution graph, we can tell that younger people are more likely to smoke, and also more likely to be alive.

From the salary distribution graph, we can see smokers tend to have a bit more salary on average than the non-smokers. 

Therefore salary and age could act as confounding variables that might influence both smoking status and life expectancy. 

No casual inference could be made between smoking and life expectancy due to the endogeneity problem!
